```python
def compute_prop_duplicates(asset_database):
    d={}
    for year in asset_database.year.unique():
        df=asset_database.copy()
        if 'activity' in df.columns:
            df = df.dropna(subset=['activity'])
            df=df.loc[asset_database['activity']!=0]
            
        else:
            df = df.dropna(subset=['capacity'])
            df=df.loc[asset_database['capacity']!=0]
        d[year]=len(df)/len(asset_database)
    return d
    
    
```
